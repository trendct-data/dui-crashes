```python
crash = pd.read_csv("data/export_25955/export_25955_0.csv",skiprows=1)
crash.head()
```

```python
def to_date(dtstr,hrstr="00:00:00"):
    return datetime.strptime(dtstr + " " + hrstr, '%m/%d/%Y %H:%M:%S')

crash["DATETIME"] = crash.apply(lambda x: to_date(x["Date Of Crash"],x["Time Of Crash"]),axis=1)
```

```python
crash["DATETIME"].head()
```

### Table 1. All crashes by hour of day



```python
def by_hour_of_day(df):
    return df.groupby(df["DATETIME"].map(lambda x: x.strftime("%H:00")) ).agg({"Crash ID":np.size})

by_hour_of_day(crash)
```

### Table 2. All crashes by day of week

### Table 3. Fatal crashes by day of week

```python
def by_day_of_week(df):
    return df.groupby(df["DATETIME"].map(lambda x: x.weekday())).agg({"Crash ID":np.size})

by_day_of_week(fatals)
```

### Table 4. Fatality/All accidents (ratio) by day of week


```python
by_day_of_week(fatals) / by_day_of_week(crash)
```

### Table 5. Fatal crashes by hour of day

```python
by_hour_of_day(fatals)
```

### Table 6. Fatal crash/all crashes (ratio) by hour of day


```python
by_hour_of_day(fatals) / by_hour_of_day(crash).sort_values(by="Crash ID",ascending=False)
```

# Table 7. Fatal crashes by day of the year

```python

def by_day_of_year(df):
    return df.groupby(df["DATETIME"].map(lambda x: x.strftime("%b %d"))).agg({"Crash ID":np.size})

by_day_of_year(fatals).sort_values(by="Crash ID",ascending=False).head(50)
```

# Data point: Median fatal crashes per day of year over 20 years

```python
by_day_of_year(fatals).median()
```

# Data point: Mean fatal crashes per day of year over 20 years

```python
by_day_of_year(fatals).mean()
```

# Table 8. Number of crashes over 20 year period by day of year

```python
by_day_of_year(crash).sort_values(by="Crash ID",ascending=False).head(50)
```

# Table 11. Most crashes by hour of year

```python
def by_day_and_hour(df):
    return df.groupby(df["DATETIME"].map(lambda x: x.strftime("%b %d %H"))).agg({"Crash ID":np.size})

by_day_and_hour(crash).sort_values(by="Crash ID",ascending=False).head(50)
```
